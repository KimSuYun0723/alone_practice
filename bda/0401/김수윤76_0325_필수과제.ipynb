```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#결측치 제거(겹치는 것들, 결측치가 너무 작은 것, 결측치가 너무 큰 것)
df= df.drop('embark_town', axis=1)
df= df.drop('deck', axis=1)
df= df.drop('class', axis=1)
df= df.drop('alive', axis=1)
df= df.drop('who', axis=1)
```

```python
df
```

```python
#결측치 중간값으로 대체
df_median=df.age.median()
df['age']=df['age'].fillna(df_median)
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder #문자형 데이터를 숫자로 바꾸는 작업
pd.get_dummies(df['sex'])
```

```python
le=LabelEncoder()
le.fit(df['sex'])
df['sex']=le.transform(df['sex']) #남성은 1, 여성은 0으로 표현된다.
```

```python
le=LabelEncoder()
le.fit(df['embarked'])
df['embarked']=le.transform(df['embarked']) #2는 Southampton, 1은 Cherbourg, 0은 Queenstown를 나타낸다.
```

```python
le=LabelEncoder()
le.fit(df['adult_male'])
df['adult_male']=le.transform(df['adult_male']) #1은 True, 0은 False를 나타낸다.
```

```python
le=LabelEncoder()
le.fit(df['alone'])
df['alone']=le.transform(df['alone']) #1은 True, 0은 False를 나타낸다.
```
